Creating a line chart.

Formatting the chart:

 - Adding titles and labels,
 - Setting y-axis limits,
 - Rotating x-axis labels.
 
 Then displaying the plot.

```python
plt.plot(indices, memory_values_in_gb, color='blue', marker='o', linestyle='-', linewidth=2, markersize=5)

plt.title('Total Memory Consumption')
plt.xlabel('Variables')
plt.ylabel('Memory in GB')

plt.ylim(0, 14)

plt.xticks(rotation=45)

plt.grid(True)
plt.tight_layout()  # Adjust layout to avoid clipping of labels
plt.show()
```
