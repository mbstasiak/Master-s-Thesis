Display basic information about the DataFrame
(4467307 rows, 62 columns)

```python
info_df = pd.DataFrame(dataframe_v01.dtypes, columns=['Data Type'])
info_df['Non-Null Count'] = dataframe_v01.count()
info_df['Missing Values Count'] = dataframe_v01.isnull().sum()
info_df['Non-Missing Values Percentage'] = (info_df['Non-Null Count'] / len(dataframe_v01)) * 100
info_df['Unique Count'] = dataframe_v01.nunique()
info_df.reset_index(inplace=True)
info_df = info_df.rename(columns={'index': 'Column Name'})
print("DataFrame Info:")
info_df
```

Display statistical information about the DataFrame

```python
# Display statistical information about the DataFrame
print("\nDataFrame Description:")
dataframe_v01.describe(include='all')
```

Create a new DataFrame to store columns with less than 55 unique values and their unique values

```python
unique_values_df = pd.DataFrame(columns=['Column Name', 'Unique Values'])

# Populate the DataFrame with columns having less than 55 unique values
for column in dataframe_v01.columns:
    if dataframe_v01[column].nunique() < 55:
        unique_values = dataframe_v01[column].unique()
        unique_values_df = unique_values_df.append({'Column Name': column, 'Unique Values': unique_values}, ignore_index=True)

print("Columns with Less Than 55 Unique Values:")
unique_values_df

# The number of unique values in the column "Property State" is 54.
```

Create a new DataFrame for columns with less than 80% non-missing values


```python
columns_less_than_80_df = info_df.loc[info_df['Non-Missing Values Percentage'] < 80, ['Column Name', 'Non-Missing Values Percentage']]

print("\nDataFrame of Columns with Less Than 80% Non-Missing Values:")
columns_less_than_80_df
```
